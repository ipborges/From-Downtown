###Estruturando os dados

Para realizar uma análise temporal mais robusta, os dados foram reestruturados utilizando a linguagem R. A partir do conjunto original `team_stats_game`, foi criado um novo conjunto denominado `stats_season`. Nesse processo, foram selecionadas as variáveis relevantes, filtrando apenas os dados a partir da temporada de 1979.

A estruturação envolveu as seguintes etapas:

1. **Seleção de Variáveis:** Foram escolhidas as variáveis-chave relacionadas a arremessos (médias por jogo e porcentagens) e a temporada.

2. **Filtragem por Temporada:** Foram considerados apenas os dados a partir da temporada de 1979, ano marcante com a introdução da linha de 3 pontos na NBA.

3. **Agrupamento e Resumo:** Os dados foram agrupados por temporada e, em seguida, calculadas médias para as variáveis selecionadas, resultando em um conjunto de dados mais condensado e focado na evolução temporal dos arremessos.

```r
# @title
stats_season <- team_stats_game %>%
  select(fg_per_game, fga_per_game, fg_percent, x2p_per_game, x2pa_per_game, x2p_percent, x3p_per_game, x3pa_per_game, x3p_percent, season) %>%
  filter(as.numeric(season) > 1979) %>%
  group_by(season) %>%
  summarise(
    avg_fg_per_game = mean(fg_per_game),
    avg_fg_percent = mean(fg_percent),
    avg_x2p_per_game = mean(x2p_per_game),
    avg_x2pa_per_game = mean(x2pa_per_game),
    avg_x2p_percent = mean(x2p_percent),
    avg_x3p_per_game = mean(x3p_per_game),
    avg_x3pa_per_game = mean(x3pa_per_game),
    avg_x3p_percent = mean(x3p_percent)
  )
```

### **3. A Mudança de Perfil dos Pivôs**

Utilizei consultas SQL no bigquery para aprofundar o entendimento sobre a mudança de perfil nas posições 4 (pivôs) e 5 (alas-pivôs). Posições que desde a o príncipio da NBA possuem caractéristicas de um jogo mais dentro do garrafão, a area mais interna da quadra.As consultas SQL registradas e executadas foram as seguintes:

```r
# @title
projeto <- "sunny-diorama-410222"
dataset <- "nba_3pt_evolution"

arremessos_123 <- 'CREATE TABLE `sunny-diorama-410222.nba_3pt_evolution.pg_sg_sf_shooting` AS
SELECT
  season,
  player,
  x3pa,
  x3p_percent,
  pos
FROM
  `sunny-diorama-410222.nba_3pt_evolution.player_totals`
WHERE
  pos = "PG" OR pos = "SG" OR pos = "SF";'


resultado_arremessos_123 <- bq_project_query(projeto, arremessos_123)


arremessos_45 <- 'CREATE TABLE `sunny-diorama-410222.nba_3pt_evolution.center_powerforward_shooting` AS
SELECT
  season,
  player,
  x3pa,
  x3p_percent,
  pos
FROM
  `sunny-diorama-410222.nba_3pt_evolution.player_totals`
WHERE
  pos = "C" OR pos = "PF";'


resultado_arremessos_45 <- bq_project_query(projeto, arremessos_45)


ev_arremessos_45 <- 'SELECT
  season,
  SUM(x3pa) AS total_x3pa
FROM `sunny-diorama-410222.nba_3pt_evolution.center_powerforward_shooting`
WHERE
  season != 2024
GROUP BY
  season
ORDER BY
  total_x3pa DESC;'


resultado_ev_arremessos_45 <- bq_project_query(projeto, ev_arremessos_45)


dados_ev_arremessos_45 <- bq_table_download(resultado_ev_arremessos_45)


ev_arremessos_123 <- 'SELECT
  season,
  SUM(x3pa) AS total_x3pa
FROM `sunny-diorama-410222.nba_3pt_evolution.pg_sg_sf_shooting`
WHERE
  season != 2024
GROUP BY
  season
ORDER BY
  total_x3pa DESC;'


resultado_ev_arremessos_123 <- bq_project_query(projeto, ev_arremessos_123)


dados_ev_arremessos_123 <- bq_table_download(resultado_ev_arremessos_123)


print("Evolução do número de arremessos pelos pivôs:")
print(dados_ev_arremessos_45)

print("Evolução do número de arremessos dos jogadores das posições 1, 2 e 3:")
print(dados_ev_arremessos_123)
```

Essa observação destaca uma mudança significativa no papel dos pivôs na NBA ao longo do tempo. Anteriormente, os pivôs eram frequentemente associados ao jogo no garrafão, com habilidades mais voltadas para a área próxima à cesta e menos enfatizando arremessos de longa distância, especialmente de três pontos.


```r
# @title
dif_percentual <-'WITH Summary45 AS (
  SELECT
    season,
    SUM(x3pa) AS total_x3pa_45
  FROM
    nba_3pt_evolution.center_powerforward_shooting
  WHERE
    season != 2024
  GROUP BY
    season
),

Summary123 AS (
  SELECT
    season,
    SUM(x3pa) AS total_x3pa_123
  FROM
    nba_3pt_evolution.pg_sg_sf_shooting
  WHERE
    season != 2024
  GROUP BY
    season
)

SELECT
  COALESCE(a.season, b.season) AS season,
  COALESCE(total_x3pa_45, 0) AS total_x3pa_45,
  COALESCE(total_x3pa_123, 0) AS total_x3pa_123,
  COALESCE((total_x3pa_123 - total_x3pa_45) / NULLIF(total_x3pa_123, 0) * 100, 0) AS percentage_difference
FROM
  Summary45 a
FULL JOIN
  Summary123 b
ON
  a.season = b.season
ORDER BY
  total_x3pa_45 DESC;'

resultado_dif_percentual <- bq_project_query(projeto, dif_percentual)


dados_dif_percentual <- bq_table_download(resultado_dif_percentual)


print("Diferença percentual entre o número de arremessos:")
print(dados_dif_percentual)
```

Ao comparar os últimos anos, notamos uma evolução significativa no estilo de jogo dos pivôs, evidenciada pelo aumento da contribuição dos arremessos de três pontos. Em 2014, a diferença no número de arremessos em comparação com os demais jogadores era de 82,4%. A partir de 2017, essa diferença permaneceu na faixa dos 60%, chegando a atingir 57,3% no ano de 2020. Vale destacar que, ao compararmos as posições de pivô (ala-pivô e pivô) com as demais (armadores, ala-armadores e alas), representamos "40% dos jogadores" pensando somente nas 5 posições em quadra. Portanto, uma diferença próxima aos 33% sugeriria que os pivôs estão arremessando proporcionalmente tanto quanto os jogadores das outras posições, como armadores, ala-armadores e alas.





# Considerações Finais e Insights


Ao analisar profundamente a evolução dos arremessos de três pontos na NBA, fica evidente que esse aspecto do jogo passou por uma transformação significativa. Entre 1995 e 1997, surgiram grandes arremessadores como Reggie Miller e Dennis Scott, que deixaram um impacto marcante com seus recordes. No entanto, o salto verdadeiramente expressivo ocorreu a partir de 2013, impulsionado pelo sucesso das equipes como Heat, Spurs e Warriors. As demais equipes, buscando replicar o que funcionou para esses times, adotaram a estratégia de arremessos de longa distância.

O declínio nos arremessos de dois pontos durante esse período é notável, indicando que os arremessos de três pontos não são mais apenas um artifício no jogo, mas tornaram-se a estratégia principal adotada pelas equipes. Essa mudança de paradigma destaca a importância de eficiência nos arremessos para alcançar o sucesso. Arremessos excessivos, sem qualidade, não foram suficientes para conduzir equipes a títulos ou finais.

Outro ponto de destaque é a transformação no perfil dos pivôs. Tradicionalmente focados no jogo no garrafão, os pivôs passaram a contribuir mais efetivamente nos arremessos, acompanhando a evolução do jogo.

Na NBA contemporânea, a bola de três pontos tornou-se um fator chave para a vitória. Estratégias em quadra devem priorizar arremessos de longa distância com alta efetividade, uma vez que arremessos excessivos, sem eficiência, demonstraram ser contraproducentes. Além disso, jogadores em todas as posições não podem mais prescindir da habilidade de serem bons arremessadores de três pontos. Em um cenário atual, a versatilidade no jogo de arremesso é crucial para o sucesso das equipes na NBA.
